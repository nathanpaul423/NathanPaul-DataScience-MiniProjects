```python
df.groupby('License Type')['Email'].count().sort_values(ascending=False)
```

```python
df_filter1 = df[df['License Type'].isin(['Broker','Managing Broker',"Principal Broker"])]
```

```python
df_filter1
```

```python
df_office_county = df_filter1.groupby('Office County')['Email'].count().sort_values(ascending=False).to_frame().reset_index()
```

```python
df_office_county.rename(columns={'Office County':'County','Email':'Count'},inplace=True)
```

```python
df_office_county_filter = df_office_county[(df_office_county['Count'] >= 15) & (df_office_county['Count'] <= 100)]
```

```python
county = df_office_county_filter['County'].to_list()
county
```

```python
df_filter1[df_filter1['Office County'].isin(county)]
```
