```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# example 1

X = range(100)
Y = [value ** 2 for value in X]
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# example 2
T = range(100)
X = [(2*math.pi*t)/ len(T) for t in T]
Y = [math.sin(value) for value in X]
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# example 3  same as example 2 but using numpy

X = np.linspace(0,2* np.pi, 100)
Y = np.sin(X)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# example 4
X = np.linspace(-3,2,200)
Y = X**2-2*X+1
plt.plot(X,Y)
plt.show()
```

```python
# example 5
X = np.linspace(0,2*np.pi,100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X,Ya)
plt.plot(X,Yb)
plt.show()
```

```python
#example 6
def plot_slope(X,Y):
  Xs = X[1:] - X[:-1]
  Ys = Y[1:] - Y[:-1]
  plt.plot(X[1:],Ys/Xs)

X = np.linspace(-3,3,100)
Y = np.exp(-X**2)

plt.plot(X,Y)
plot_slope(X,Y)

plt.show()
```
