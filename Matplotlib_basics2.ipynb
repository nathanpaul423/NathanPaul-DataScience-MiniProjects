```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# example 1

X = range(100)
Y = [value ** 2 for value in X]
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# example 2
T = range(100)
X = [(2*math.pi*t)/ len(T) for t in T]
Y = [math.sin(value) for value in X]
```

```python
plt.plot(X,Y)
plt.show()
```

```python
# example 3  same as example 2 but using numpy

X = np.linspace(0,2* np.pi, 100)
Y = np.sin(X)
```

```python
plt.plot(X,Y)
plt.show()
```
